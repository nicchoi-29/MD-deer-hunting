```python
#Lyme Disease 
#This dataset is by county 
import pandas as pd

#1. Read CSV
df_plyme = pd.read_csv("df_plyme.csv")
```

```python
#2. Subset update dataframe to locate only Maryland's incidence rates and create new datframe just for MD

#     #df_plyme[df_plyme['STNAME'] == 'Maryland'].index
#             #12474:72332
#             #non-consecutive
#             #552 rows total

#create new df called maryland
df_plyme_updated=df_plyme[df_plyme['STNAME'] == 'Maryland']
```

```python
#Lyme Disease 

#3. Change incidence rate so that all tickborne diseases are per 1000000 people (lyme is 100,000)
    # Need to first group by county and year
    # Then calculate totals for cases and population 
df_maryland_lyme_totals = df_plyme_updated.groupby("YEAR")[["CTYNAME", "CASES", "POP"]].sum()
```

```python
#3. Continued
# Then need to change rate:



df_maryland_lyme_totals['ONE_MILLION_RATE'] = (df_maryland_lyme_totals["CASES"] / df_maryland_lyme_totals["POP"]) * 1000000
```

```python
#Lyme Disease

#4. Subset MD dataframe to extract incidence rates only from 2015-2022 (to somewhat match harvest data timeline which goes up to 2025)
df_maryland_lyme_totals=df_maryland_lyme_totals.iloc[14:]
```
