```python
#Spotted Fever Rickettsiosis (SFR)

#1. Read CSV 

import pandas as pd 
df_psfr = pd.read_csv("df_psfr.csv")
df_psfr
```

```python
#Spotted Fever Rickettsiosis (SFR)

#2. Change incidence rate so that all tickborne diseases are per 1000000 people 
df_psfr['ONE_MILLION_RATE'] = (df_psfr["CASES"] / df_psfr["POP"]) * 1000000

#tried 100,000 but rates were too small to readily see a noticeable pattern. 
```

```python
#Spotted Fever Rickettsiosis (SFR)

#3. Drop "RATE" column from original table and create new dataframe 

df_psfr_updated = df_psfr.drop(["RATE"], axis = 1)
```

```python
#Spotted Fever Rickettsiosis (SFR)

#4. Subset update dataframe to locate only Maryland's incidence rates and create new datframe just for MD

    #RangeIndex(start=240, stop=252, step=1)
    #https://www.statology.org/pandas-get-row-number/

df_psfr_updated=df_psfr_updated[df_psfr_updated['NAME'] == 'Maryland']
```

```python
df_psfr_updated
```

```python
#Spotted Fever Rickettsiosis (SFR)

#5. Subset MD dataframe to extract incidence rates only from 2015-2022 (to somewhat match harvest data timeline which goes up to 2025)

#2015 begins at row 475 

df_psfr_updated = df_psfr_updated.drop(df_psfr_updated.loc[:474].index)                        
```

```python
#Spotted Fever Rickettsiosis (SFR)

#6. Drop columns that don't need to be in chart

df_psfr_updated = df_psfr_updated.drop(["STATEFP", "STUSPS", "NAME", "CASES", 
"POP"], axis = 1)
```

```python
#Spotted Fever Rickettsiosis (SFR)
#7. Reset Index 
df_psfr_updated=df_psfr_updated.reset_index(drop=True)
```

```python
#Spotted Fever Rickettsiosis (SFR)
#8. Add disease column for grouped bar chart 
df_psfr_updated=df_psfr_updated.assign(TICKBORNE_DISEASE = "Spotted Fever Rickettsiosis")
#https://stackoverflow.com/questions/44657226/how-to-insert-a-new-column-with-repeated-values-into-a-pandas-table
```

```python
df_psfr_updated.to_csv("df_psfr_updated.csv", index=False)

#gather incidence rates in millions for other diseases in maryland 2015-2025 and do stacked bar chart
```
